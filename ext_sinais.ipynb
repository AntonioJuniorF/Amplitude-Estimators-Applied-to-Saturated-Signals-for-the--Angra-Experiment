# Importando os dados

```python
def open_cvc(filename):
    with h5py.File(filename, 'r') as f:
        # List all groups
        print("Keys: %s" % f.keys())
        a_group_key = list(f.keys())[0]

        # Get the data
        Sinais = list(f[a_group_key])
        Sinais = np.array(Sinais)
    return Sinais
```

```python
plt.figure(figsize=[10,8])
for i in range(209):
  plt.plot(S_inp[i][0])
plt.title('Vendo alguns sinais de entrada',fontsize = 20)
plt.xlabel('Time (samples)', fontsize = 16)
plt.ylabel('Amplitude (Volts)', fontsize = 16)
#plt.xlim(0,300)
plt.grid()
plt.show()
```

```python
def ped_sinais(Sinais) : # está função cálcula o valor dos pedestais dos canais com base dos 100 primeiros pontos de cada sinal
    ped = np.zeros((Sinais.shape[0],Sinais.shape[1]))
    for i in range(Sinais.shape[0]):
        ped[i] = np.mean(Sinais[i][:,0:150],axis =  1)
    return ped
```

```python
def remov_ped(Sinais): # esta função remove o pedestal dos sinais 
    ped = ped_sinais(Sinais) # Cálcula o pedestal dos sinais 
    S = np.zeros((Sinais.shape[0],Sinais.shape[1],Sinais.shape[2]))
    for i in range(Sinais.shape[0]):
        S[i] = Sinais[i] - ped[i].reshape(Sinais.shape[1],1)
    return S
```

```python
S_inp = remov_ped(S_inp) # removendo o pedestal do sinal do laboratório
```

```python
plt.figure(figsize=[10,8])
for i in range(209):
  plt.plot(S_inp[i][0])

plt.title('Sinal de entrada com o pedestal removido',fontsize = 20)

plt.xlabel('Time (samples)', fontsize = 16)
plt.ylabel('Amplitude (Volts)', fontsize = 16)
#plt.xlim(0,300)
plt.grid()
plt.show()
```

### Amplitude de pico
---
 
- No caso da amplitude de pico será considerada o maior valor em módulo dos sinais.

```python
def amplitude(S):
  #Amp = np.zeros((S.shape[0],S.shape[1]))
  #for i in range(S.shape[0]):
  Amp =  np.max(S,axis = 1)   
  return Amp
```

```python
def amplitude_1(S):
  Amp = np.zeros((S.shape[0],S.shape[1]))
  for i in range(S.shape[0]):
    Amp[i] =  np.min(S[i],axis = 1)   
  return Amp
```

### área
 
---
 
* A área dos sinais será calculada pelo somatório das amostras vezes o intervalo que cada amostra foi adquirido, no caso 8 ns.


```python
def carga1(S):
    f = 1.25*10**9
    T = 1/f
    #T = 8*10**-9
    area = np.zeros((S.shape[0],S.shape[1]))
    for i in range(S.shape[0]):
        area[i] =  np.sum(S[i]*T,axis = 1)
           
    return area


def carga(S):
    f = 1.25*10**9
    T = 1/f
    #T = 8*10**-9
    area =  np.sum(S*T,axis = 1)       
    return area
```

```python
Area_out   = carga(S_out)
Area_inp   = carga1(S_inp)
```

```python
Area_out = Area_out*10**9
Area_inp = Area_inp*10**9
```

### Largura à meia altura
---
* O cálculo da largura meia altura foi calculado pela primeira amostra que está acima do valor da amplitude de pico dividido por 2. Depois que este valor foi encontrado ele será dividido por dois. 
 
 
 



```python
def med_meia_a (S,Amp):
    dist = np.zeros(S.shape[0])
    for j in range(S.shape[0]):
            #print(j)
        #for i in range(Amp.shape[1]):
            ind        =  np.argmax(S[j])
            mei        =  np.abs(S[j] - (Amp[j]/2))
            #print(ind)
            #print(mei[ind:-1])
            #print(j)
            if ind <49:
                aux        = np.abs(mei[ind:-1])
             #   print(j)
             #   print(aux)
                dist[j] = 2*np.argmin(aux)
                #print(j)
            else:
                dist[j] = 0
    return dist
```

```python
Amp_out1    = amplitude(S_out)  # amplitude saída
Meia_altura = med_meia_a (S_out,Amp_out1)
```
